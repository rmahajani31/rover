```python
import csv
from mcap_ros2.reader import read_ros2_messages
```

```python
import os

bags_dir = "/home/rmahajani/Documents/projects/rover/ros2_ws/bags"
bag_name = "test_run_1"
run_idx = 0
input_mcap = os.path.join(bags_dir, bag_name, f"{bag_name}_{run_idx}.mcap")
output_csv = os.path.join(bags_dir, bag_name, f"cmd_vel_analysis_{run_idx}.csv")

print(f"Reading from {input_mcap}...")

with open(output_csv, 'w', newline='') as f:
    writer = csv.writer(f)
    # Write the header row
    writer.writerow(['timestamp_ns', 'linear_x', 'linear_y', 'angular_z'])

    # Iterate through messages in the MCAP file
    for msg in read_ros2_messages(input_mcap):
        if msg.channel.topic == '/cmd_vel':
            # msg.ros_msg is the actual geometry_msgs/Twist object
            t = msg.publish_time_ns
            lx = msg.ros_msg.twist.linear.x
            ly = msg.ros_msg.twist.linear.y
            az = msg.ros_msg.twist.angular.z
            
            writer.writerow([t, lx, ly, az])

print(f"Done! Data saved to {output_csv}")
```
